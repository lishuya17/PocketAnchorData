```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

### pocket detection

```python
# demo data
with open('lists/pdbid_chain_list_scPDB.txt') as f:
    list_task = [line.strip() for line in f.readlines()]
print(list_task[:10])
```

```python
list_sampled = np.random.choice(list_task, 100, replace=False)
len(list_sampled)
```

```python
fw = open('global_da4.sh', 'w')
for item in list_sampled:
    fw.write('python src/0_define_anchor_one_sample.py {} {}\n'.format(item, 4))
fw.close()
```

```python
fw = open('global_da6.sh', 'w')
for item in list_sampled:
    fw.write('python src/0_define_anchor_one_sample.py {} {}\n'.format(item, 6))
fw.close()
```

```python
fw = open('global_da8.sh', 'w')
for item in list_sampled:
    fw.write('python src/0_define_anchor_one_sample.py {} {}\n'.format(item, 8))
fw.close()
```

```python
num_anchors_da4 = []
for item in list_sampled:
    anchors = np.load('./AnchorOutput/{}_da_{}/anchors.npy'.format(item,4))
    num_anchors_da4.append(anchors.shape[1])

num_anchors_da6 = []
for item in list_sampled:
    try:
        anchors = np.load('./AnchorOutput/{}_da_{}/anchors.npy'.format(item,6))
        num_anchors_da6.append(anchors.shape[1])
    except Exception as E:
        print(E)

num_anchors_da8 = []
for item in list_sampled:
    anchors = np.load('./AnchorOutput/{}_da_{}/anchors.npy'.format(item,8))
    num_anchors_da8.append(anchors.shape[1])
```

```python
plt.figure(dpi=200, figsize=(10,3))
plt.subplot(1,3,1)
plt.hist(num_anchors_da4)
plt.title('da=4')
plt.ylabel('frequency')
plt.xlabel('num anchors')
plt.subplot(1,3,2)
plt.hist(num_anchors_da6)
plt.title('da=6')
plt.xlabel('num anchors')

plt.subplot(1,3,3)
plt.hist(num_anchors_da8)
plt.title('da=8')
plt.xlabel('num anchors')
```

### cpi

```python
# demo data
with open('lists/pdbid_chain_center_list_PDBbind.txt') as f:
    list_task = [line.strip() for line in f.readlines()]
print(list_task[:10])
```

```python
list_sampled = np.random.choice(list_task, 100, replace=False)
len(list_sampled)
```

```python
  for da in [2,4,6,8]:
    fw = open('local_da{}.sh'.format(da), 'w')
    for item in list_sampled:
        pdbid, center = item.split()
        fw.write('python src/0_define_anchor_one_sample_CPI.py {} {} {}\n'.format(pdbid, center, da))
    fw.close()
```

```python
num_anchors_da2 = []
for item in list_sampled:
    pdbid, center = item.split()
    anchors = np.load('./AnchorOutput/{}_center_{}_da_{}/anchors.npy'.format(pdbid, center,2))
    num_anchors_da2.append(anchors.shape[1])
    
num_anchors_da4 = []
for item in list_sampled:
    pdbid, center = item.split()
    anchors = np.load('./AnchorOutput/{}_center_{}_da_{}/anchors.npy'.format(pdbid, center,4))
    num_anchors_da4.append(anchors.shape[1])

num_anchors_da6 = []
for item in list_sampled:
    pdbid, center = item.split()
    anchors = np.load('./AnchorOutput/{}_center_{}_da_{}/anchors.npy'.format(pdbid, center,6))
    num_anchors_da6.append(anchors.shape[1])

num_anchors_da8 = []
for item in list_sampled:
    pdbid, center = item.split()
    anchors = np.load('./AnchorOutput/{}_center_{}_da_{}/anchors.npy'.format(pdbid, center,8))
    num_anchors_da8.append(anchors.shape[1])
```

```python
plt.figure(dpi=200, figsize=(5,3))
# plt.subplot(1,4,1)
plt.boxplot([num_anchors_da2,num_anchors_da4,num_anchors_da6,num_anchors_da8])
plt.ylabel('num anchors')
plt.xticks(np.arange(4)+1, [2,4,6,8])
plt.xlabel('da')
```
